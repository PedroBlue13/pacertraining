```python

formulas = {
    "Markup_simples": [
        "preco_venda = custo_produto √ó (1 + markup_percentual)"
    ],
    "Markup_inverso": [
        "preco_venda = custo_produto / (1 - margem_desejada)"
    ],
    "Markup_sobre_custos_totais": [
        "preco_venda = (custos_fixos + custos_variaveis + despesas) √ó (1 + markup_percentual)"
    ],
    "Markup_dinamico": [
        "preco_venda = custo_produto √ó fator_dinamico",
        "# onde fator_dinamico varia conforme estoque, demanda, sazonalidade etc."
    ],
    "Markup_competitivo": [
        "preco_venda = preco_medio_mercado ¬± ajuste_percentual",
        "# usado quando se baseia nos concorrentes"
    ],
    "Markup_financeiro_completo": [
        "preco_venda = custo_produto / (1 - (despesas_fixas + despesas_variaveis + Margem_lucro))"
    ],
    "Margem_contribuicao": [
        "Margem_contribuicao = preco_venda - (custos_variaveis + despesas_variaveis)"
    ],
    "Venda_margem_de_lucro_bjetiva": [
        "preco_venda = custos_produto / (1 - taxas_de_venda - margem_objetiva)"
    ],
    "Precificacao_baseada_concorrencia": [
        "preco_venda = preco_concorrente ¬± ajuste_percentual"
    ],
    "Precificacao_valor_percebido": [
        "preco_venda = valor_percebido_pelo_cliente - desconto_estrategico"
    ]
}

"""
## Estrat√©gias de Precifica√ß√£o

### So, ONI values during 1950-1955 will be based on the 1936-1965 base period, ONI values during 1956-1960 will be based on the 1941-1970 base period, and so on and so forth.

In real-time operations, the past 30-year base period (e.g. 1991‚Äì2020) will continue to be used to compute the departure from average. However, CPC will create an additional 30-year base period every 5 years (the next update will be at the beginning of 2026). When these 5 year updates occur, the ONI values over the most recent decade will change slightly because of the inclusion of more recent data.

This new strategy has two distinct advantages:

(1) The classification of El Ni√±o and La Ni√±a episodes will remain fixed over most of the historical record. So, future adjustments to the base period will not modify the past classification of episodes. For example, for the set of years between 2001‚Äì2005, the ONI values (and El Ni√±o and La Ni√±a episodes) will permanently be calculated as the departure from the 1986‚Äì2015 base period.

(2) A centered 30-year base period means that El Ni√±o and La Ni√±a episodes will be defined by their contemporary climatology. For example, the La Ni√±a episode that occurred in mid-1950s will have negative ONI values that are representative of the climatology at that time and not some future climatology.

The monthly Ni√±o-3.4 index (which is used to calculate the ONI values) that uses these new centered 30-year base periods ("ClimAdjust") are provided in this text file: üîπ Markup_simples
**Uso recomendado:** √ötil para produtos de baixo custo e venda recorrente  
**O que √©:** Aplica um percentual fixo sobre o custo do produto para obter o pre√ßo de venda. √â r√°pido e f√°cil, mas pode n√£o considerar todas as despesas indiretas.

---

### üîπ Markup_inverso
**Uso recomendado:** Ideal para quando se conhece o pre√ßo final desejado e quer calcular o custo m√°ximo aceit√°vel  
**O que √©:** Calcula o pre√ßo de venda com base no custo e na margem de lucro desejada, considerando a margem como uma fra√ß√£o do pre√ßo final.

---

### üîπ Markup_sobre_custos_totais
**Uso recomendado:** Recomendado para neg√≥cios que incluem todos os custos (fixos e vari√°veis) no pre√ßo final  
**O que √©:** Soma todos os custos envolvidos no produto e aplica um markup sobre esse total, garantindo que despesas e lucros estejam cobertos.

---

### üîπ Markup_dinamico
**Uso recomendado:** Indicado para mercados sazonais ou com demanda vari√°vel  
**O que √©:** O fator de markup varia conforme crit√©rios como estoque, sazonalidade, urg√™ncia de venda ou comportamento do consumidor.

---

### üîπ Markup_competitivo
**Uso recomendado:** Estrat√©gia √∫til em mercados com muita concorr√™ncia e pre√ßos transparentes  
**O que √©:** Define o pre√ßo com base no valor praticado pelos concorrentes, ajustando conforme o posicionamento da marca.

---

### üîπ Markup_financeiro_completo
**Uso recomendado:** Ideal para precifica√ß√£o detalhada com despesas fixas, vari√°veis e margem de lucro  
**O que √©:** M√©todo mais robusto de markup, que calcula o pre√ßo considerando todos os custos e a margem desejada, muito usado em planilhas empresariais.

---

### üîπ Margem_contribuicao
**Uso recomendado:** Importante para analisar a lucratividade de cada produto e otimizar a gest√£o de custos  
**O que √©:** Indica quanto cada produto contribui para cobrir os custos fixos e gerar lucro. Muito √∫til para tomada de decis√£o e mix de produtos.

---

### üîπ Venda_margem_de_lucro_bjetiva
**Uso recomendado:** Ideal para produtos com custos vari√°veis e margem de lucro definida  
**O que √©:** Define o pre√ßo com base nos custos, taxas e na margem de lucro desejada. Boa escolha para empresas que buscam controle financeiro rigoroso.

---

### üîπ Precificacao_baseada_concorrencia
**Uso recomendado:** Indicada para mercados competitivos, onde a compara√ß√£o de pre√ßos √© comum  
**O que √©:** Analisar os pre√ßos praticados pelos concorrentes e definir um pre√ßo similar ou ligeiramente diferente, dependendo do posicionamento da marca e diferenciais do produto.

---

### üîπ Precificacao_valor_percebido
**Uso recomendado:** Ideal para produtos com alto valor agregado e diferenciais competitivos  
**O que √©:** Define o pre√ßo com base no valor que o cliente percebe no produto ou servi√ßo, considerando diferenciais como exclusividade, qualidade, experi√™ncia de uso, etc.
"""
```

```python
from pprint import pprint

def evaluating_criteria(low_cost, recurrent_sales, know_final_price, high_value_added, high_competition, seasonality, financial_control):
    strategies = []

    if low_cost and recurrent_sales:
        strategies.append("Markup_simples")

    if know_final_price:
        strategies.append("Markup_inverso")

    if not low_cost and financial_control:
        strategies.append("Markup_financeiro_completo")

    if seasonality:
        strategies.append("Markup_dinamico")

    if high_competition:
        strategies.append("Precificacao_baseada_concorrencia")
        strategies.append("Markup_competitivo")

    if high_value_added:
        strategies.append("Precificacao_valor_percebido")

    if financial_control and not know_final_price:
        strategies.append("Venda_margem_de_lucro_objetiva")

    if not recurrent_sales and not high_competition:
        strategies.append("Markup_sobre_custos_totais")

    strategies.append("Margem_contribuicao")

    return list(set(strategies))  



market_section_characteristics = {
    "Fresh Produce": {
        "low_cost": True,
        "recurrent_sales": True,
        "know_final_price": False,
        "high_value_added": False,
        "high_competition": True,
        "seasonality": True,
        "financial_control": True
    },
    "Bakery": {
        "low_cost": False,
        "recurrent_sales": True,
        "know_final_price": False,
        "high_value_added": True,
        "high_competition": True,
        "seasonality": False,
        "financial_control": True
    },
    "Beverages": {
        "low_cost": True,
        "recurrent_sales": True,
        "know_final_price": False,
        "high_value_added": False,
        "high_competition": True,
        "seasonality": False,
        "financial_control": True
    },
    "Cleaning & Household": {
        "low_cost": True,
        "recurrent_sales": True,
        "know_final_price": True,
        "high_value_added": False,
        "high_competition": True,
        "seasonality": False,
        "financial_control": True
    }
}


section_pricing_strategies = {}

for section, param in market_section_characteristics.items():
    section_pricing_strategies[section] = evaluating_criteria(**param)


pprint(section_pricing_strategies)
```

```python
import numpy as np
import pandas as pd

# Carregar o dataset
df = pd.read_csv("pricing_dataset.csv")

# Engenharia de atributos
df["delta_preco"] = df["preco_unit"].diff()
df["media_movel"] = df["vendas"].rolling(window=3).mean()

# Como "semana" √© apenas um n√∫mero, vamos simular uma data fict√≠cia para calcular o dia da semana
df["data_ficticia"] = pd.date_range(start="2023-01-01", periods=len(df), freq="W")
df["dia_da_semana"] = df["data_ficticia"].dt.dayofweek

# Sele√ß√£o de vari√°veis (usando preco_unit e vendas)
X = df[["preco_unit"]].values
y = df["vendas"].values

# Regress√£o linear simples
x_mean = X.mean()
y_mean = y.mean()

w = ((X - x_mean) * (y - y_mean)).sum() / ((X - x_mean)**2).sum()
b = y_mean - w * x_mean

y_pred = w * X + b

# Exibir coeficientes
print(f"Coeficiente angular (w): {w:.4f}")
print(f"Intercepto (b): {b:.2f}")
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

mse = mean_squared_error(y, y_pred)
mae = mean_absolute_error(y,y_pred)

print(f"MSE: {mse:.2f}, MAE: {mae:.2f}")
```
